# Data Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
data = pd.read_csv(url)
print(data)
```

```python
X = data['Hours'].values
Y = data['Scores'].values
```

# Simple Visualizing Student Data

```python
plt.scatter(X,Y)
plt.title("Hours vs Scores")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.plot([f(x) for x in range(600)], c='red')
plt.show()
```

# Calculating Linear Regression Parameters Using SciPy's linregress Function

```python
from scipy import stats
a, b, r, p, se = stats.linregress(X, Y)
print(a,b,r,se)
```

# Visualizing Student Data along with the linear regression line


```python
def f(x) :
    y= a*x+b
    return y

plt.scatter(X,Y)
plt.title("Hours vs Scores")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.plot([f(x) for x in range(11)], c='red')
plt.show()
```

# Training and Testing Data Split

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X1 = data['Hours'].values.reshape(-1, 1)
y1 = data['Scores'].values

X_train, X_test, Y_train, Y_test = train_test_split(X1, Y1, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(X_train, Y_train)
```

# Model Evaluation Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

# Score Prediction for 9.25 Hours of Study

```python
hours = 9.25
predicted_score = model.predict([[hours]])
print(f"Predicted score for {hours} hours of study: {predicted_score[0]}")
```
